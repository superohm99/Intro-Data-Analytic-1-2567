```python
df = pd.read_csv(CSV_PATH)
```

```python
df
```

```python
df.groupby('CustomerID')['InvoiceNo'].nunique().sort_values(ascending=False).head(5)
```

```python
df.groupby('CustomerID')['UnitPrice'].head(5)
```

```python
df["Expense"] = df["Quantity"] * df["UnitPrice"]
```

```python
df
```

```python
df.groupby(['CustomerID','Country'])['Expense'].sum().reset_index()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['Hour'] = df['InvoiceDate'].dt.hour
```

```python
df
```

```python
df['TotalCost'] = df['Quantity'] * df['UnitPrice']
```

```python
df.groupby('Hour')['TotalCost'].sum().reset_index()
```

```python
december_sales = df[df['InvoiceDate'].dt.month == 12]
```

```python
december_sales['TotalSales'] = december_sales['Quantity']
```

```python
december_sales.groupby('StockCode')['TotalSales'].sum().reset_index().sort_values(by="TotalSales",ascending=False).head(5)
```
